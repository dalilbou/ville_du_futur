```python
import requests
from faker import Faker
import random
from datetime import timedelta
import unidecode

# Initialize Faker
fake = Faker('fr_FR')

# Step 1: Make the API call to get real addresses
response = requests.get("https://api-adresse.data.gouv.fr/search/?q=Châtenay-Malabry&postcode=92290&limit=100")
data = response.json()

# Step 2: Function to select a real address and add a random number
def fetch_address_from_data(data):
    street_number = random.randint(1, 10)  # Generate a random street number between 1 and 10
    if data['features']:
        selected_address = random.choice(data['features'])
        # Add the random street number to the beginning of the address
        return f"{street_number} {selected_address['properties']['label']}"
    else:
        # Fallback to a fake address with the random number
        return f"{street_number} {fake.street_address()}, 92290 Châtenay-Malabry"

# Step 3: Generate a fake user with real address
def generate_fake_user(data):
    first_name = fake.first_name()
    last_name = fake.last_name()
    normalized_first_name = unidecode.unidecode(first_name.lower().replace(' ', '_'))
    normalized_last_name = unidecode.unidecode(last_name.lower().replace(' ', '_'))
    email = f"{normalized_first_name}.{normalized_last_name}@example.com"
    start_date = fake.date_between(start_date='-30d', end_date='today')
    end_date = start_date + timedelta(days=random.randint(2, 6))  # Randomly set an end date between 2 and 6 days after start_date

    # Generate a phone number
    phone = f"06{random.randint(10000000, 99999999)}"

    # Fetch a real or fallback fake address with a random number
    address = fetch_address_from_data(data)

    # Example user dictionary
    user_data = {
        'first_name': first_name,
        'last_name': last_name,
        'email': email,
        'phone': phone,
        'address': address,
        'start_date': start_date,
        'end_date': end_date
    }

    return user_data

# Step 4: Create and display 5 test users
test_users = [generate_fake_user(data) for _ in range(5)]
for user in test_users:
    print(user)
```
